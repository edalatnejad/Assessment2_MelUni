```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error

# Importing the non_emty column from Weather.csv in a DataFram(df_weather) 
selected_Col=['Date','Minimum temperature (°C)','Maximum temperature (°C)','Rainfall (mm)','Direction of maximum wind gust ','Speed of maximum wind gust (km/h)','Time of maximum wind gust','9am Temperature (°C)','9am relative humidity (%)','9am wind direction','9am wind speed (km/h)','9am MSL pressure (hPa)',	'3pm Temperature (°C)','3pm relative humidity (%)',	'3pm wind direction','3pm wind speed (km/h)','3pm MSL pressure (hPa)']
df_weather= pd.read_csv('weather.csv',usecols=selected_Col)

# Importing all columns from price_and_demand.csv in a DataFram(def_Consuming) 
def_Consuming = pd.read_csv('price_and_demand.csv')

# split SETTLEMENTDATE column in two columns(Date and Time)
def_Consuming[['Date','Time']]=def_Consuming['SETTLEMENTDATE'].str.split(expand=True)
# Rmove un_nessesory columns
def_Consuming = def_Consuming.drop(columns=['SETTLEMENTDATE','REGION','RRP','PERIODTYPE','Time'])
# Find mean of TOTALDamand for every Date
def_Consuming_groupedBy=def_Consuming.groupby('Date')['TOTALDEMAND'].max()
# Merge two df
result_df = pd.merge(df_weather, def_Consuming_groupedBy, on='Date', how='inner')

#################### Fill NaN data
result_df['Maximum temperature (°C)'].fillna(result_df['Maximum temperature (°C)'].mean(), inplace=True)
result_df['3pm Temperature (°C)'].fillna(result_df['3pm Temperature (°C)'].mean(), inplace=True)
result_df['3pm relative humidity (%)'].fillna(result_df['3pm relative humidity (%)'].mean(), inplace=True)
result_df['3pm MSL pressure (hPa)'].fillna(result_df['3pm MSL pressure (hPa)'].mean(), inplace=True)
result_df['9am MSL pressure (hPa)'].fillna(result_df['9am MSL pressure (hPa)'].mean(), inplace=True)
result_df['Speed of maximum wind gust (km/h)'].fillna(result_df['Speed of maximum wind gust (km/h)'].mean(), inplace=True)
result_df['3pm wind speed (km/h)'].fillna(result_df['3pm wind speed (km/h)'].mean(), inplace=True)
result_df['Direction of maximum wind gust '].fillna('SSW', inplace=True) # because We have 63 record with SSW (The max count)

#result_df
```

```python
Light rain — when the precipitation rate is < 2.5 mm (0.098 in) per hour
Moderate rain — when the precipitation rate is between 2.5 mm (0.098 in) – 7.6 mm (0.30 in) or 10 mm (0.39 in) per hour[108][109]
Heavy rain — when the precipitation rate is > 7.6 mm (0.30 in) per hour,[108] or between 10 mm (0.39 in) and 50 mm (2.0 in) per hour[109]
```

```python
def categorize_rainfall(rainfall):
    if rainfall < 2.5:
        return 'Light'
    elif 2.5 <= rainfall <= 7.6 or 10 <= rainfall <= 50:
        return 'Moderate'
    elif rainfall > 7.6:
        return 'Heavy'
    else:
        return 'Unknown'
```

```python
def categorize_windSpeed(wind_speed_kph):
    if wind_speed_kph < 1:
        return  "Calm"
    elif 1 <= wind_speed_kph < 6:
        return  "Light Air"
    elif 6 <= wind_speed_kph < 12:
        return  "Light Breeze"
    elif 12 <= wind_speed_kph < 20:
        return "Gentle Breeze"
    elif 20 <= wind_speed_kph < 29:
        return  "Moderate Breeze"
    elif 29 <= wind_speed_kph < 39:
        return  "Fresh Breeze"
    elif 39 <= wind_speed_kph < 50:
        return  "Strong Breeze"
    elif 50 <= wind_speed_kph < 62:
        return  "Moderate Gale"
    elif 62 <= wind_speed_kph < 75:
        return "Fresh Gale or Gale"
    elif 75 <= wind_speed_kph < 89:
        return  "Strong Gale"
    elif 89 <= wind_speed_kph < 103:
        return  "Storm"
    elif 103 <= wind_speed_kph < 118:
        return  " Violent Storm"
    else:
        return  "Hurricane"
    
    # Reference:https://sciencing.com/classify-wind-speeds-23181.html
```

```python
def apply_Rainfall_categorization():
    result_df['Rainfall'] =result_df['Rainfall (mm)'].apply(categorize_rainfall)
    return result_df['Rainfall']
   
    
```

```python
def apply_windSpeed_categorization():
    result_df['MaxWindSpeed'] =result_df['Speed of maximum wind gust (km/h)'].apply(categorize_windSpeed)
    return result_df['MaxWindSpeed']
```
