```python
import pandas as pd
import matplotlib.pyplot as plt

# Importing the non_emty column from Weather.csv in a DataFram(df_weather) 
selected_Col=['Date','Minimum temperature (°C)','Maximum temperature (°C)','Rainfall (mm)','Direction of maximum wind gust ','Speed of maximum wind gust (km/h)','Time of maximum wind gust','9am Temperature (°C)','9am relative humidity (%)','9am wind direction','9am wind speed (km/h)','9am MSL pressure (hPa)',	'3pm Temperature (°C)','3pm relative humidity (%)',	'3pm wind direction','3pm wind speed (km/h)','3pm MSL pressure (hPa)']
df_weather= pd.read_csv('weather.csv',usecols=selected_Col)
# Importing all columns from price_and_demand.csv in a DataFram(def_Consuming) 

def_Consuming = pd.read_csv('price_and_demand.csv')
# split SETTLEMENTDATE column in two columns(Date and Time)
def_Consuming[['Date','Time']]=def_Consuming['SETTLEMENTDATE'].str.split(expand=True)
# Rmove un_nessesory columns
def_Consuming = def_Consuming.drop(columns=['SETTLEMENTDATE','REGION','RRP','PERIODTYPE','Time'])
# Find mean of TOTALDamand for every Date
def_Consuming_groupedBy=def_Consuming.groupby('Date')['TOTALDEMAND'].max()
# Merge two df
result_df = pd.merge(df_weather, def_Consuming_groupedBy, on='Date', how='inner')
#print(result_df.head(10000))
```

```python
#'Direction of maximum wind gust ','9am wind direction','3pm wind direction','Time of maximum wind gust'
FEATURES = ['TOTALDEMAND','Minimum temperature (°C)','Maximum temperature (°C)','Speed of maximum wind gust (km/h)','9am Temperature (°C)','9am relative humidity (%)','9am wind speed (km/h)','9am MSL pressure (hPa)','3pm Temperature (°C)','3pm relative humidity (%)',	'3pm wind speed (km/h)','3pm MSL pressure (hPa)']
# Calculate the Pearson's R correlation metric
result_df[FEATURES].corr(method='pearson')
```

```python
plt.scatter(y=result_df['TOTALDEMAND'],x=result_df['Maximum temperature (°C)'], alpha=0.5)
plt.xlabel('Maximum temperature (°C)')
plt.ylabel('TOTALDEMAND')
plt.show()
#*************************************************
plt.scatter(y=result_df['TOTALDEMAND'],x=result_df['Minimum temperature (°C)'], alpha=0.5)
plt.xlabel('Minimum temperature (°C)')
plt.ylabel('TOTALDEMAND')
plt.show()
#*************************************************
plt.scatter(y=result_df['TOTALDEMAND'],x=result_df['9am Temperature (°C)'], alpha=0.5)
plt.xlabel('9am Temperature (°C)')
plt.ylabel('TOTALDEMAND')
plt.show()
#*************************************************
plt.scatter(y=result_df['TOTALDEMAND'],x=result_df['3pm Temperature (°C)'], alpha=0.5)
plt.xlabel('3pm Temperature (°C)')
plt.ylabel('TOTALDEMAND')
plt.show()
#*************************************************
plt.scatter(y=result_df['TOTALDEMAND'],x=result_df['3pm relative humidity (%)'], alpha=0.5)
plt.xlabel('3pm relative humidity (%)')
plt.ylabel('TOTALDEMAND')
plt.show()
```
